```python
#파라미터 튜닝
X_train = train_data.name
y_train = train_data.middle_classification
X_test = test_data.name
y_test = test_data.middle_classification

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
pipeline = Pipeline([
                     ('vectorizer',TfidfVectorizer(min_df = 0.0, analyzer='char', sublinear_tf=True)),
                     ('lgs', LogisticRegression(class_weight='balanced'))
])

params ={
    'vectorizer__ngram_range':[(1,1),(1,2)],
    'vectorizer__max_df':[50000,100000],
    'lgs__C':[1,5,10]
}

grid_cv_pipe = GridSearchCV(pipeline,param_grid=params,cv=2,scoring='accuracy',verbose=1)
grid_cv_pipe.fit(X_train,y_train)
print(grid_cv_pipe.best_params_,grid_cv_pipe.best_score_)

pred = grid_cv_pipe.predict(X_test)
print('Pipeline을 통한 LR의 예측 정확도는{0:3f}'.format(accuracy_score(y_test,pred)))
#print("Accuracy: %f" % lgs.score(X_eval, y_eval))
```
